```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12,8)
```

```python
data = pd.read_csv('bike-dataset.txt')
print(data.head())
```

```python
print(data.info())
```

```python
ax = sns.scatterplot('Population', 'Profit', data = data)
ax.set_title('profit in $10000 vs Popualation in 10000s');
```

```python
x = data['Population'].values
y = data['Profit'].values

mean_x = np.mean(x)
mean_y = np.mean(y)

numer = demon = 0
numer = np.sum((x - mean_x)*(y - mean_y))
denom = np.sum((x - mean_x) ** 2)
m  = numer / denom
c  = mean_y - mean_x * m

print(m ,c)
```

```python

X = np.linspace((min(x) - 1 ),(max(x)+1), 1000)
y_pre = (m*X) +c

plt.plot(X, y_pre, color = 'b')
plt.scatter(x, y, color='r')
```
